## Part 2: Plotting US and Italy Data

We can extract the data for the US from the dataframe

```python
df_us = df[df['Country/Region'] == 'US']
df_us = df_us.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_us = df_us.drop(columns=['Lat','Long','Province/State'])

data_us = df_us.loc['US','2/28/20':'4/5/20']
```

## 2a: Plot Italy and US data in two axes

Use subplots and assign separate colors to each country. Set the same limits on the y-axis for both subplots

```python
# use 
fig,ax = plt.subplots(1,2,figsize=(15,5)) # or whatever dimensions you like
ax[0].plot(data_italy.values,'.-',color='red',linewidth=3)
ax[0].set_title('Covid in Italy')
ax[1].plot(data_us.values,'.-',color='blue',linewidth=3)
ax[1].set_title('Covid in US')

for a in ax:
    a.set_xlabel('Days')
    a.set_ylabel('Cases')
    a.set_ylim([0,350000])
```

## 2b: Plot Italy and US data in the same plot

Use the same colors as above, and include a legend

```python
# your code here
plt.plot(data_italy.values,'.-',color='red',linewidth=3,label='covid in italy')
plt.plot(data_us.values,'.-',color='blue',linewidth=3, label='covid in US')
plt.legend()
plt.show()
```

## 2c Plot Italy and US data on log axis

Note that we only want the y-axis to be logarithmic

```python
import numpy as np
# your code here
plt.plot(np.log(data_italy.values),'.-',color='red',linewidth=3,label='covid in italy')
plt.plot(np.log(data_us.values),'.-',color='blue',linewidth=3, label='covid in US')
plt.ylabel('Log of Cases')
plt.xlabel('Days')
plt.legend()
plt.show()
```

## 2d Make the same plot as above, but using a `for` loop

```python
# your code here
data = pd.concat([data_italy,data_us],axis=1)


for i in (data.keys()):
    plt.plot(np.log(data[i]),label = i)
    plt.legend()
```

## Part 3 Plot Data for 5 countries

Use a for loop, and include x/y labels and a legend. Also save your figure as a jpg and share with friends. You are now a Python datascientist

```python
countries = ['US','Norway','Iceland','Sweden','Japan']

for i in countries:
    df_sub = df[df['Country/Region'] == i]
    df_sub = df_sub.set_index('Country/Region', drop = True) # set pandas dataframe index to country
    df_sub = df_sub.drop(columns=['Lat','Long','Province/State']) 
    data_sub = df_sub.loc[i,:]
    plt.plot(np.log(data_sub),label = i)
    plt.legend()
    plt.xticks(rotation=90)

# save your figure
plt.savefig('test.jpg')
```
